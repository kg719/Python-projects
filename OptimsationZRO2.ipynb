```python
import numpy as np

x = np.linspace(0.5,0.99,500)
tabR = []
for i in (x):
    XZr = 1/((3*i)+(5*(1-i)))
    XY = 2*(1-i)*(XZr)
    r = XZr/XY
    tabR.append(r)
error = 76
for i in tabR:
    if abs(i-6)< error:
        error = abs(i-6)
        ValueX = (x[tabR.index(i)])
print(tabR)
print(ValueX)
```
